Dropping Unnecessary Columns from COVID-19 Data

```python
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)
```

Displaying First 5 Rows of Modified COVID-19 Data

```python
covid_df.head()
```

Converting 'Date' Column to DateTime Format

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

Displaying First 5 Rows of Data with DateTime Format

```python
covid_df.head()
```


Calculating Active Cases

```python
covid_df['Active_cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

Creating Statewise Pivot Table for COVID-19 Data

```python
statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc="max")
```

Calculating Recovery Rate for Each State/Union Territory

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

Calculating Mortality Rate for Each State/Union Territory

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

Sorting States by Confirmed Cases

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

Applying Background Gradient to Statewise Data

```python
statewise.style.background_gradient(cmap = "rainbow")
```

Identifying Top 10 States with Highest Active Cases

```python
top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by = ['Active_cases'], ascending = False).reset_index()
```

Creating Bar Plot for Top 10 States with Most Active Cases

```python
top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by = ['Active_cases'], ascending = False).reset_index()

fig = plt.figure(figsize = (16, 9))

plt.title("Top 10 states with most active cases in India", size = 25)

ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'red')

plt.xlabel("states")
plt.ylabel("Total active cases")
plt.show()
```

Top states with Highest deaths

```python
top_10_deaths = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by='Deaths', ascending=False).reset_index()

fig = plt.figure(figsize = (18,5))

plt.title("Top 10 states with most Deaths", size = 25)

ax = sns.barplot(data = top_10_deaths.iloc[:12], y = "Deaths", x = "State/UnionTerritory", linewidth = 2, edgecolor = "black")

plt.xlabel("states")
plt.ylabel("Total Death cases")
plt.show()
```

Plotting Trends in Confirmed, Cured, Deaths, and Active Cases Over Time in India

```python
india_data = covid_df.groupby('Date').sum()[['Confirmed', 'Cured', 'Deaths', 'Active_cases']]
plt.figure(figsize=(14, 7))
plt.plot(india_data.index, india_data['Confirmed'], label="Confirmed Cases", color='blue')
plt.plot(india_data.index, india_data['Cured'], label="Cured Cases", color='green')
plt.plot(india_data.index, india_data['Deaths'], label="Deaths", color='red')
plt.plot(india_data.index, india_data['Active_cases'], label="Active Cases", color='orange')
plt.title("Covid-19 Trends in India Over Time", size=20)
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.legend()
plt.show()
```


Plotting Mortality Rate Trend Over Time in India

```python
india_data['Mortality Rate'] = (india_data['Deaths'] / india_data['Confirmed']) * 100

plt.figure(figsize=(12, 6))
plt.plot(india_data.index, india_data['Mortality Rate'], label="Mortality Rate", color='red')
plt.title("Mortality Rate Trend in India", size=20)
plt.xlabel("Date")
plt.ylabel("Mortality Rate (%)")
plt.grid()
plt.show()
```


Plotting Case Doubling Rate Trend Over Time in India

```python
india_data['Doubling Rate (Days)'] = india_data['Confirmed'].diff().rolling(window=7).apply(
    lambda x: (np.log(2) / np.log(x.iloc[-1] / x.iloc[0])) if x.iloc[0] > 0 and x.iloc[-1] > 0 else np.nan
)
india_data = india_data.dropna(subset=['Doubling Rate (Days)'])

plt.figure(figsize=(12, 6))
plt.plot(india_data.index, india_data['Doubling Rate (Days)'], label="Doubling Rate", color='purple')
plt.title("Case Doubling Rate Trend in India", size=20)
plt.xlabel("Date")
plt.ylabel("Doubling Rate (Days)")
plt.grid(True)
plt.xticks(rotation=45)  # Rotate date labels for better readability
plt.legend()
plt.show()
```

Forecasting Future COVID-19 Confirmed Cases Using ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA

# Ensure that the index has a daily frequency
india_data.index = pd.to_datetime(india_data.index)
india_data = india_data.asfreq('D', method='pad')  # Fill missing dates if necessary

# Predicting Future Cases Using ARIMA
confirmed_cases = india_data['Confirmed'].dropna()
model = ARIMA(confirmed_cases, order=(5, 1, 0))  # Adjust the ARIMA order as needed
model_fit = model.fit()

# Forecast for the next 30 days
forecast = model_fit.forecast(steps=30)
forecast_index = pd.date_range(start=india_data.index[-1], periods=31, freq='D')[1:]

# Plotting the historical data and forecasted values
plt.figure(figsize=(14, 7))
plt.plot(confirmed_cases, label="Historical Data", color='blue')
plt.plot(forecast_index, forecast, label="Forecast", color='orange')
plt.title("Covid-19 Confirmed Cases Forecast", size=20)
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.legend()
plt.show()
```
